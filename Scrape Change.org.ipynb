```python
##Write function to get appropriate URL
def getURL(ID):

    URLpart1 = "https://www.change.org/api-proxy/-/petitions/"

    finalURL = URLpart1 +str(ID)
    
    return(finalURL)
```

```python
def checkEmpty(ID):
    #request the URL
    request = getURL(ID)
    request = req = urllib2.Request(getURL(ID), headers={'User-Agent' : "Magic Browser"}) 
    #ensure url exists
    try:
        response = urllib2.urlopen( req )
        is_data=True
    except Exception:
        is_data=False
    #get data in a pandas format
    if is_data==True:
        readResponse  = response.read()
        Json_text = json.loads(readResponse);
        pd_Format = json_normalize(Json_text)
        return(pd_Format)
    else:
        return(False)


```

```python
def loop_check_for_data(Start_ID, Num_Iterations):
    #initialize variable for dataframe creation
    created_df_yet = False
    for iteration in range(Num_Iterations):
        ID = iteration+Start_ID
       # print(ID)
        request = getURL(ID)
        request = req = urllib2.Request(getURL(ID), headers={'User-Agent' : "Magic Browser"}) 
        #ensure url exists
        try:
            response = urllib2.urlopen( req )
            is_data=True
        except Exception:
            is_data=False
        #get data in a pandas format
        if is_data==True:
            readResponse  = response.read()
            Json_text = json.loads(readResponse);
            pd_Format = json_normalize(Json_text)
            #Append data if dataframe not yet initialized
            if created_df_yet==True:
                pd_data = pd_data.append(pd_Format)
            else:
                pd_data=pd_Format
                created_df_yet=True
        
    #set index to id
    pd_data.index = pd_data.id
    print(ID)
    return(pd_data)
```
